```python
import os
import shutil
import subprocess
from pathlib import Path
from urllib.request import urlretrieve

REPO_URL = "https://github.com/nikuradzedima/dl_bdz.git"
CHECKPOINT_URL = "https://huggingface.co/dimadmitrij734/bdz_first/resolve/main/model_best.pth"
AUDIO_URL = "https://keithito.com/LJ-Speech-Dataset/LJ025-0076.wav"
REPO_DIR = Path("/content/dl_bdz")
CHECKPOINT_PATH = Path("checkpoints/model_best.pth")
OUTPUT_PATH = "outputs/demo_reconstructed.wav"
DEVICE = "cuda"
```

```python
shutil.rmtree(REPO_DIR, ignore_errors=True)
subprocess.run(["git", "clone", REPO_URL, str(REPO_DIR)], check=True)

os.chdir(REPO_DIR)
subprocess.run(["python", "-m", "pip", "install", "-q", "-r", "requirements.txt"], check=True)
print("cwd:", Path.cwd())
```

```python
CHECKPOINT_PATH.parent.mkdir(parents=True, exist_ok=True)
urlretrieve(CHECKPOINT_URL, CHECKPOINT_PATH)

print("checkpoint:", CHECKPOINT_PATH)
print("audio url:", AUDIO_URL)
```

```python
cmd = [
    "python",
    "inference.py",
    f"checkpoint_path={CHECKPOINT_PATH}",
    f"input_url={AUDIO_URL}",
    f"output_path={OUTPUT_PATH}",
    f"device={DEVICE}",
]
subprocess.run(cmd, check=True)
```

```python
from IPython.display import Audio, display

print("Original")
display(Audio("outputs/input_from_url.wav"))
print("Reconstructed")
display(Audio(OUTPUT_PATH))
```
